```python
##  Country / Top 10 confirmed Deaths
## Check Again


Sorted_Confirmed_Deaths = df.sort_values(by='Deaths', ascending = False)

Sorted_Confirmed_Deaths_top10 = Sorted_Confirmed_Deaths.head(10).sort_values(by='Deaths', ascending=False)

plt.figure(figsize=(10,6))
plt.bar(Sorted_Confirmed_Deaths_top10['Country/Region'], Sorted_Confirmed_Deaths_top10['Deaths'], color='red')

plt.xlabel('Country/Region', fontsize=12)
plt.ylabel('Confirmed Deaths', fontsize=12)
plt.title('Top 10 Countries Ordered by Confirmed Deaths', fontsize=14)
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.tight_layout()
plt.show()

## Country / Amount of Confirmed Deaths, Sorted
Sorted_Confirmed_Deaths = df.sort_values(by='Deaths', ascending = False)

plt.figure(figsize=(30, 8))
plt.bar(Sorted_Confirmed_Deaths['Country/Region'], Sorted_Confirmed_Deaths['Deaths'], color ='red')
               
plt.xlabel('Country/Region', fontsize=12)
plt.ylabel('Confirmed Deaths (M)', fontsize=12)
plt.title('Countries Ordered by Confirmed Deaths', fontsize=14)
plt.xticks(rotation=90)  # Rotate country names for better visibility
plt.tight_layout()
```

```python
## Top 10 Countries New Cases

# Group the data by 'Country/Region' and sum the columns
global_trends = df.groupby('Country/Region')[['New cases', 'New deaths', 'New recovered']].sum()

# Now you can use global_trends for analysis
top_10_new_cases = global_trends['New cases'].sort_values(ascending=False).head(10)

# Select the data for the top 10 countries
top_10_data = global_trends.loc[top_10_new_cases.index]

# Plot the top 10 countries' new cases, new deaths, and new recoveries as horizontal bars
ax = top_10_data.plot(kind='barh', figsize=(12, 8), title="Top 10 Countries by New COVID-19 Cases, Deaths, and Recoveries", 
                      color=['orange', 'red', 'green'])

# Set labels and title
plt.xlabel("Number of Cases", fontsize=12)
plt.ylabel("Country/Region", fontsize=12)
plt.tight_layout()  # Adjust layout to prevent label overlap

# Display the plot
plt.show()

#######

## Top 10 countries with most death percentage from new cases

# Corrected Death Rate calculation (new deaths divided by new cases)
global_trends['Death Rate (%)'] = (global_trends['New deaths'] / global_trends['New cases']) * 100

# Sort the data by the 'Death Rate (%)' and select the top 10 countries
top_10_death_rate = global_trends['Death Rate (%)'].sort_values(ascending=False).head(10)

# Select the data for the top 10 countries
top_10_data_death_rate = global_trends.loc[top_10_death_rate.index]

# Plot the death rate as a horizontal bar plot
top_10_data_death_rate['Death Rate (%)'].plot(kind='barh', figsize=(12, 8), color='red', title="Top 10 Countries by Death Rate (%)")
plt.xlabel("Death Rate (%)", fontsize=12)
plt.ylabel("Country/Region", fontsize=12)
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```

```python
# Group the data by 'WHO Region' and sum the columns
global_trends = df.groupby('WHO Region')[['New cases', 'New deaths', 'New recovered']].sum()

# Now you can use global_trends for analysis
top_10_new_cases = global_trends['New cases'].sort_values(ascending=False).head(10)

# Select the data for the top 10 WHO Regions
top_10_data = global_trends.loc[top_10_new_cases.index]

# Plot the top 10 WHO Regions' new cases, new deaths, and new recoveries as horizontal bars
ax = top_10_data.plot(kind='barh', figsize=(12, 8), title="Top 10 WHO Regions by New COVID-19 Cases, Deaths, and Recoveries", 
                      color=['orange', 'red', 'green'])

# Set labels and title
plt.xlabel("Number of Cases", fontsize=12)
plt.ylabel("WHO Region", fontsize=12)
plt.tight_layout()  # Adjust layout to prevent label overlap

# Display the plot
plt.show()

#######

## Top 10 WHO Regions with most death percentage from new cases

# Corrected Death Rate calculation
global_trends['Death Rate (%)'] = (global_trends['New deaths'] / global_trends['New cases']) * 100  # Divide new deaths by new cases

# Sort the data by the 'Death Rate (%)' and select the top 10 WHO regions
top_10_death_rate = global_trends['Death Rate (%)'].sort_values(ascending=False).head(10)

# Select the data for the top 10 WHO regions
top_10_data_death_rate = global_trends.loc[top_10_death_rate.index]

# Plot the death rate as a horizontal bar plot
top_10_data_death_rate['Death Rate (%)'].plot(kind='barh', figsize=(12, 8), color='red', title="Top 10 WHO Regions by Death Rate (%)")
plt.xlabel("Death Rate (%)", fontsize=12)
plt.ylabel("WHO Region", fontsize=12)
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()

```

```python
# Sort by the '1 week % increase' column and select the top 10 countries with the highest increase
top_10_weekly_increase = df.sort_values(by='1 week % increase', ascending=False).head(10)

# Plot the top 10 countries' 1-week percentage increase as a bar plot
plt.figure(figsize=(12, 8))
plt.barh(top_10_weekly_increase['Country/Region'], top_10_weekly_increase['1 week % increase'], color='blue')

# Set labels and title
plt.xlabel("1 Week % Increase", fontsize=12)
plt.ylabel("Country/Region", fontsize=12)
plt.title("Top 10 Countries by 1-Week % Increase in New COVID-19 Cases", fontsize=14)
plt.tight_layout()

# Display the plot
plt.show()

```

```python
# Group by 'WHO Region' and calculate the sum of '1 week % increase' for each region
who_region_data = df.groupby('WHO Region')['1 week % increase'].sum()

# Sort the regions by the '1 week % increase' in descending order and select the top 10
top_10_who_regions = who_region_data.sort_values(ascending=False).head(10)

# Plot the top 10 WHO regions' 1-week percentage increase as a bar plot
plt.figure(figsize=(12, 8))
plt.barh(top_10_who_regions.index, top_10_who_regions.values, color='blue')

# Set labels and title
plt.xlabel("1 Week % Increase", fontsize=12)
plt.ylabel("WHO Region", fontsize=12)
plt.title("Top 10 WHO Regions by 1-Week % Increase in New COVID-19 Cases", fontsize=14)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Calculate the death ratio based on new deaths and 1 week % increase in new cases
top_10_who_regions['Death Ratio'] = top_10_who_regions['New deaths'] / (top_10_who_regions['1 week % increase'] / 100)

# Set up a data frame for the heatmap, including the new death ratio
heatmap_data = top_10_who_regions[['New deaths', '1 week % increase', 'Death Ratio']]

plt.figure(figsize=(12, 8))

# Using the 'Reds' colormap for a red-toned heatmap
sns.heatmap(heatmap_data.T, annot=True, cmap='Reds', cbar=True, fmt='.2f', linewidths=0.5)

# Title and x-axis label
plt.title("Heatmap of Top 10 WHO Regions by Week Case % Increase, New Deaths, and Death Ratio", fontsize=14)
plt.xlabel("WHO Regions", rotation=0
, fontsize=12)

# Rotate the region names on the y-axis
plt.yticks(rotation=45, fontsize=12)  # Adjust rotation if needed

plt.tight_layout()
plt.show()



```
